````python
#
# ---- Importing the modules ----
#

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
````

````python
#
# ---- Reading the datasets ----
#

spotify_data = pd.read_csv('./datas/spotify.csv'
                           , index_col='Date'
                           , parse_dates=True)

flight_data = pd.read_csv('./datas/flight_delays.csv'
                         , index_col='Month')

insurance_data = pd.read_csv('./datas/insurance.csv')

iris_data = pd.read_csv('./datas/iris.csv'
                      , index_col='Id')
````

----

<h1 id='trend-plots' style='color:blue'>0) Trend Plots</h1>

<br>

<p><b>Trend Plots</b> are used to show patterns of changes, especially over the time, such as hours, days, weeks, years and so on. Styles:</p>

* Line Plots

<h2 id='line-plots' style='color:blue'>0.1) Line Plots</h2>

<br>

<p><b>Line Plots</b> are usefull to show changes of variables over the time. Commands:</p>

<br>

> All Variables

```
sns.lineplot(data=data)
```

<br>

> Just a Few Variables

```
sns.lineplot(data=data1, label='label1')
sns.lineplot(data=data2, label='label2)
```

````python
#
# ---- Showing a line plot with all columns ----
#

plt.figure(figsize=(10,7))
plt.title('Songs on Spotify: Number of Views per Date')
plt.xlabel('Date')
plt.ylabel('Number of Views')

sns.lineplot(data=spotify_data)

plt.show()
````

````python
#
# ---- Line Plot with just a few columns/variables ----
#

plt.figure(figsize=(10,7))

plt.title('Spotify Songs: Number of Views per Date')
plt.xlabel('Date')
plt.ylabel('Number of Views')

sns.lineplot(data=spotify_data['Despacito'], label='Despacito')
sns.lineplot(data=spotify_data['Something Just Like This'], label='Something Just Like This')
sns.lineplot(data=spotify_data['Unforgettable'], label='Unforgettable')

plt.show()
````

----

<h1 id='relationship-plots' style='color:blue'>1) Relationship Plots</h1>

<br>

<p><b>Relationship Plots</b> are used to show comparisons/relationships between the dataset variables. Styles:</p>

* Bar Plot
* HeatMap Plot
* Scatter Plot
* Regression Plot
* Lm Plot
* Swarm Plot

<h2 id='bar-plots' style='color:blue'>1.0) Bar Plots</h2>

<br>

<p><b>Bar Plots</b> compare quantities corresponding from different groups. Commands:</p>

<br>

> Simple Bar Plot

```
sns.barplot(data=data, x=x_data, y=y_data)
```

<br>

> Bar Plot with Categorical Variable

```
sns.barplot(data=data, x=x_data, y=y_data, hue=categorical variable)
```


````python
#
# ---- Simple Bar Plot ----
#

plt.figure(figsize=(10,7))
plt.title("Month Delays of HA's Flights")

# positive values >> the flight was delayed
# negative values >> the flight was early
sns.barplot(data=flight_data, x=flight_data.index, y='HA')

plt.show()
````

````python
#
# ---- Bar Plot with Categorical Variable ----
#

plt.figure(figsize=(10,7))
plt.title('Charges by BMI and Smoker or Not People')


sns.barplot(data=insurance_data[0:20]
           , x='bmi'
           , y='charges'
           , hue='smoker')

plt.show()
````

<br>

<h2 id='heatmap-plots' style='color:blue'>1.1) HeatMap Plots</h2>

<br>

<p><b>HeatMaps</b> show patterns and correlations colouring the variables, only accept numerical variables and categorical variables must be labelled. Commands:</p>

<br>

> Simple HeatMap

```
sns.heatmap(data=data, annot=True|False)
```

````python
#
# ---- Simple HeatMap ----
#

plt.figure(figsize=(10,7))

plt.title('Flights Delays per Month')
plt.xlabel('Flights')
plt.ylabel('Months')

sns.heatmap(data=flight_data, annot=True)

plt.show()
````

<br>

<h2 id='scatter-plots' style='color:blue'>1.2) Scatter Plots</h2>

<br>

<p><b>Scatter Plots</b> show how the data are distributed in the dataset. Commands:</p>

<br>

> Simple Scatter Plot

```
sns.scatterplot(data=data, x=x_data, y=y_data)
```

<br>

> Scatter Plot with Categorical Variable

```
sns.scatterplot(data=data, x=x_data, y=y_data, hue=categorical_variable)
```

````python
#
# ---- Simple Scatter Plot ----
#

plt.figure(figsize=(10,7))

plt.title('Charges by BMI')
plt.xlabel('BMI')
plt.ylabel('Charges')

sns.scatterplot(data=insurance_data, x='bmi', y='charges')

plt.show()
````

````python
#
# ---- Scatter Plot with Categorical Variable ----
#

plt.figure(figsize=(10,7))

plt.title('Charges by BMI and Smoker or Not People')
plt.xlabel('BMI')
plt.ylabel('Charges')

sns.scatterplot(data=insurance_data, x='bmi', y='charges', hue='smoker')

plt.show()
````

<br>

<h2 id='regression-plots' style='color:blue'>1.3) Regression Plots</h2>

<br>

<p><b>Regression Plots</b> are like the scatter ones, but with a regression line displaying the relationship between the x and y variables. Commands:</p>

<br>

> Regression Plot

```
sns.regplot(data=data, x=x_data, y=y_data)
```

````python
#
# ---- Regression Plot ----
#

plt.figure(figsize=(10,7))

plt.title('Charges by BMI')
plt.xlabel('BMI')
plt.ylabel('Charges')

# as higher the BMI is, as more the person pays for charges
sns.regplot(data=insurance_data, x='bmi', y='charges')

plt.show()
````

<br>

<h2 id='swarm-plots' style='color:blue'>1.4) LM Plots</h2>

<br>

<p><b>LM Plots</b> are like the regression ones, but with a categorical variable and a regression line for each categorical group. Commands:</p>

<br>

> LM Plot

```
sns.lmplot(data=data, x=x_data, y=y_data, hue=categorical_variable)
```

````python
#
# ---- LM Plot ----
#

# we can assume that if the person is a smoker, they will pay
# for more charges as higher the BMI is.
#
# the same is valid for non-smokers, but the charge value is way
# to slow compared to the smoker ones
sns.lmplot(data=insurance_data, x='bmi', y='charges', hue='smoker');
````

<br>

<h2 id='swarm-plots' style='color:blue'>1.5) Swarm Plots</h2>

<br>

<p><b>Swarm Plots</b> show the relationship between a categorical variable and a numerical one. Commands:</p>

<br>

> Swarm Plot

```
sns.swarmplot(data=data, x=x_data, y=y_data)
```

````python
#
# ---- Swarm Plot ----
#

plt.figure(figsize=(10,7))

plt.title('Charges by Smoker or Non-Smoker People')
plt.xlabel('Smoker or Not')
plt.ylabel('Charges')

sns.swarmplot(data=insurance_data, x='smoker', y='charges')

plt.show()
````

<h1 id='distribution-plots' style='color:blue'>2) Distribution Plots</h1>

<br>

<p><b>Distribution Plots</b> show how the datas are distributed into the dataset and help the Data Scientist takes guesses of the possible output values for specific inputs. Styles:</p>

* Histogram Plots
* KDE Plots
* Joint Plots

<br>

<h2 id='histogram-plots' style='color:blue'>2.0) Histogram Plots</h2>

<br>

<p><b>Histogram Plots</b> show the distribution of a single a variable with or without a categorical one. Commands:</p>

<br>

> Simple Hist Plot

```
sns.histplot(data=data, x=x_data)
```

<br>

> Hist Plot with Categorical Variable

```
sns.histplot(data=data, x=x_data, hue=categorical_variable)
```

````python
#
# ---- Simple Hist Plot ----
#

plt.figure(figsize=(10,7))

plt.title('Number of Flowers by Sepal Length (cm)')
plt.xlabel('Count')
plt.ylabel('Sepal Length (cm)')

sns.histplot(data=iris_data, x='Sepal Length (cm)')

plt.show()
````

````python
#
# ---- Hist Plot with Categorical Variable ----
#
plt.figure(figsize=(10,7))

plt.title('Number of Flowers by Sepal Length (cm) and Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')

sns.histplot(data=iris_data, x='Sepal Length (cm)', hue='Species')

plt.show()
````

````python
#
# ---- Simple Histogram that I liked so badly ----
#

plt.figure(figsize=(10,7))

plt.title('Number of Flowers by Sepal Length (cm)')
plt.xlabel('Count')
plt.ylabel('Sepal Length (cm)')

sns.histplot(data=iris_data)

plt.show()
````

<br>

<h2 id='kde-plots' style='color:blue'>2.1) KDE Plots</h2>

<br>

<p><b>KDE Plots</b> show a smoother distribution (compared to the histograms) of one variable with or without a categorical one. Commands:</p>

<br>

> Simple KDE Plot

```
sns.kdeplot(data=data, x=x_data, shade=True|False)
```

<br>

> KDE Plot with Categorical Variable

```
sns.kdeplot(data=data, x=x_data, hue=categorical_variable, shade=True|False)
```

````python
#
# ---- Simple KDE Plot ----
#
plt.figure(figsize=(10,7))

plt.title('Number of Flowers per Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')

sns.kdeplot(data=iris_data, x='Sepal Length (cm)', shade=True)

plt.show()
````

````python
# KDE Plot with Categorical Variable

plt.figure(figsize=(10,7))

plt.title('Number of Flowers by Sepal Length (cm) and Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')

sns.kdeplot(data=iris_data, x='Sepal Length (cm)', hue='Species', shade=True)

plt.show()
````

<br>

<h2 id='joint-plots' style='color:blue'>2.2) Joint Plots</h2>

<br>

<p><b>Joint Plots</b> shows a KDE Plot of two numerical variables merged. Commands:</p>

<br>

> Joint Plot

```
sns.jointplot(data=data, x=x_data, y=y_data, kind='kde)
```

````python
#
# ---- Joint Plot ----
#
sns.jointplot(data=iris_data
             , x='Sepal Length (cm)'
             , y='Petal Length (cm)'
             , kind='kde')
````
